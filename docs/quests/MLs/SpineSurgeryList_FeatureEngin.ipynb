### 데이터 전처리

```python
df_SSL
```

```python
# 이상치 제거
list=['ODI', 'PI', 'PT', '골밀도', '지방축적도']
def outlier(x):
  for i in list :
    IQR = x[i].quantile(0.75)-x[i].quantile(0.25)
    min = x[i].quantile(0.25)-(1.5*IQR)
    max = x[i].quantile(0.75)+(1.5*IQR)
    return df_SSL.query(f'{min}<{i}<{max}')
```

```python
df_SSL = outlier(df_SSL)
df_SSL
```

```python
df_SSL_extract = df_SSL[['재발여부','ODI', 'PI', 'PT', '골밀도', '지방축적도', '가족력', '과거수술횟수']].copy()
df_SSL_extract.dropna(inplace=True)
df_SSL_extract.isnull().sum()
df_SSL_extract
```

#### Scaling

```python
from sklearn.preprocessing import MinMaxScaler
minMaxScaler=MinMaxScaler()
def scaler(x):
  minMaxScaler.fit(x.values.reshape(-1, 1))
  result = minMaxScaler.transform(x.values.reshape(-1, 1))
  return result
```

```python
df_SSL_extract["ODI_scale"]=scaler(df_SSL_extract["ODI"])
df_SSL_extract["ODI_scale"]
```

```python
df_SSL_extract["PI_scale"]=scaler(df_SSL_extract["PI"])
df_SSL_extract["PT_scale"]=scaler(df_SSL_extract["PT"])
df_SSL_extract["골밀도_scale"]=scaler(df_SSL_extract["골밀도"])
df_SSL_extract["지방축적도_scale"]=scaler(df_SSL_extract["지방축적도"])
```

#### Onehot Encoding

```python
from sklearn.preprocessing import OneHotEncoder
def onehot(x):
  oneHotEncoder = OneHotEncoder()
  oneHotEncoder.fit(x)
  df_encoder=pd.DataFrame(data= oneHotEncoder.transform(x).toarray(), columns=oneHotEncoder.get_feature_names_out())
  result = pd.concat((df_SSL_extract.reset_index(drop=True), df_encoder.reset_index(drop=True)), axis=1)
  return result
```

```python

df_SSL_extract=onehot(df_SSL_extract[["가족력"]])
df_SSL_extract[:2]
```

```python
df_SSL_extract=onehot(df_SSL_extract[["과거수술횟수"]])
df_SSL_extract[:2]
```

## 정형화 단계

```python
from sklearn.model_selection import train_test_split
target = df_SSL_extract['재발여부']
features=df_SSL_extract.drop(columns=['재발여부',	'ODI',	'PI',	'PT',	'골밀도',	'지방축적도',	'가족력',	'과거수술횟수' ])
train_features, test_features, train_target, test_target=train_test_split(features,target)
train_features.shape, test_features.shape, train_target.shape, test_target.shape
```

## 모델

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model
```

```python
model.fit(train_features, train_target)
```

```python
train_target_predict=model.predict(train_features)
train_target.shape , train_target_predict.shape
```

```python
test_target_predict=model.predict(test_features)
test_target.shape , test_target_predict.shape
```

## 평가 수치

### accuracy_score: 정확도

```python
from sklearn.metrics import accuracy_score
accuracy_score(train_target, train_target_predict) # train - 정확도 높음
```

```python
accuracy_score(test_target, test_target_predict) # test - 정확도 높음
```

### F1 score: 정밀도/재현율

```python
from sklearn.metrics import classification_report
print(classification_report(train_target, train_target_predict)) # macro avg / weighted avg의 간극이 큼
```

```python
print(classification_report(test_target, test_target_predict)) # macro avg / weighted avg의 간극이 큼
```
